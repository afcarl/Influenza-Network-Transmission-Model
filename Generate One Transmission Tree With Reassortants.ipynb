```python
from environment import Environment
import networkx as nx
from sampler import Sampler
from network_reconstructor import *
from numpy.random import normal
from random import random
from Bio import SeqIO
```

```python
e = Environment(num_viruses=2, virus_type='influenza')
for i in range(1, 100):
    if len(e.GetViruses()) < 10:
        cointoss = random()
        if cointoss > 0.7:
            virus = e.GetRandomVirus()
            e.ReplicateVirus(virus, date=i)
            e.MutateVirus()
        else:
            e.RandomlyReassortTwoViruses(mutate=True, date=i)

    else:
        break
        
e.GetViruses()
```

```python
s = Sampler(directory_prefix='Simulated One Reassortant')
s.SampleVirusesFromEnvironment(e, 'all')
```
