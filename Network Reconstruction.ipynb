```python
import numpy as np
import pandas as pd
import networkx as nx
```

```python
segment0 = pd.read_csv('Similarity Matrices/Model 2 Segment 0.csv', index_col=0)
segment1 = pd.read_csv('Similarity Matrices/Model 2 Segment 1.csv', index_col=0)
```

```python
np.fill_diagonal(segment0.as_matrix(), 0)
np.fill_diagonal(segment1.as_matrix(), 0)
```

```python
segment0
```

```python
segment1
```

```python
summed = segment0.as_matrix() + segment1.as_matrix()
summed = summed / 600.0
summed = np.triu(summed)
```

```python
G = nx.from_numpy_matrix(summed, create_using=nx.MultiDiGraph())
G
```

```python
G.edges(data=True)
```

```python
G.in_edges(3, data=True)
```

```python
max(edge[2]['weight'] for edge in G.in_edges(3, data=True))
```

```python
for node in G.nodes():
    edges = G.in_edges(node, data=True)
    print "edges are: %s" % edges
    if len(edges) > 0:
        max_weight = max([edge[2]['weight'] for edge in edges])
        for edge in edges:
            print "max_weight is: %s" % max_weight
            if edge[2]['weight'] < max_weight:
                print "removing edge %s->%s which has weight %s" % (edge[0], edge[1], edge[2]['weight'])
                G.remove_edge(edge[0], edge[1])
```
