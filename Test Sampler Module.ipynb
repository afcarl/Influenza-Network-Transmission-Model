```python
from environment import Environment
from sampler import Sampler
from random import random

import networkx as nx
```

```python
e = Environment(num_viruses=5)
for i in range(40):
    cointoss = random()
    if cointoss > 0.1:
        cointoss2 = random()
        if cointoss2 > 0.5:
            e.ReplicateAVirus(mutate=True)
        else:
            e.ReplicateAVirus(mutate=False)
    else:
        e.RandomlyReassortTwoViruses(mutate=False)
```

```python
researcher = Sampler()
researcher.SampleVirusesFromEnvironment(environment=e, n=20)
sampled_viruses = researcher.GetSampledViruses()
sampled_viruses
```

```python
sampled = nx.DiGraph()
```

```python
for virus in sampled_viruses:
    parent = virus.GetParent()
    child = virus.GetID()
    if parent == None:
        sampled.add_node(child)
    elif type(parent) == int:
        sampled.add_edge(parent, child, weight=1)
    elif len(parent) == 2:
        sampled.add_edge(parent[0], child, weight=0.5)
        sampled.add_edge(parent[1], child, weight=0.5)
```
