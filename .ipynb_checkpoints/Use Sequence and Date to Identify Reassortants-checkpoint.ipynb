```python
from Bio import SeqIO
from network_reconstructor import *
from networkx import read_gpickle, write_gpickle
```

```python
handle = r'Simulated Reassortant Sequences/Run 0 Simulation Segment 0 Sequences.fasta'
segment0 = [record for record in SeqIO.parse(handle, 'fasta')]
```

```python
handle = r'Simulated Reassortant Sequences/Run 0 Simulation Segment 1 Sequences.fasta'
segment1 = [record for record in SeqIO.parse(handle, 'fasta')]
```

```python
segment0s = []
for sequence in segment0:
    segment0s.append(sequence)

segment1s = []
for sequence in segment1:
    segment1s.append(sequence)
```

```python
segment0s
```

```python
G0 = nx.DiGraph()
add_nodes_from_records(G0, segment0s)
add_edges_from_records(G0, segment0s, segment=0)
print G0.nodes(data=True)
remove_incorrectly_timed_edges(G0)
matrix_values = nx.adjacency_matrix(G0).flat
remove_edges_below_threshold(G0, median(matrix_values))
nx.draw_circular(G0)
```

```python
G1 = nx.DiGraph()
add_nodes_from_records(G1, segment1s)
add_edges_from_records(G1, segment1s, segment=1)
print G1.nodes(data=True)
remove_incorrectly_timed_edges(G1)
matrix_values = nx.adjacency_matrix(G1).flat
remove_edges_below_threshold(G1, median(matrix_values))
nx.draw_circular(G1)
```

```python
G_all = add_two_graph_overlapping_edges(G0, G1)
remove_edges_not_max(G_all)
matrix_values = [value for value in nx.adjacency_matrix(G_all).flat if value != 0.0]
print matrix_values
print len(G_all.edges())
nx.draw_circular(G_all)
```
